# FUNCTIONS

```python
# Vertex class representing a point in 2D space
class Vertex:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, vertex):
        x_distance = abs(self.x - vertex.x)
        y_distance = abs(self.y - vertex.y)
        distance = np.sqrt((x_distance ** 2) + (y_distance ** 2))
        return distance

    def __repr__(self):
        return "(" + str(self.x) + "," + str(self.y) + ")"

    def getX(self):
        return self.x

    def getY(self):
        return self.y
```

```python
# Fitness class used to evaluate path's distance and score
class Fitness:
    def __init__(self, path):
        self.path = path
        self.distance = 0
        self.fitness = 0.0

    def path_distance(self):
        if self.distance == 0:
            self.distance = path_distance_using_LUT(self.path)
        return self.distance

    def path_fitness(self):
        if self.fitness == 0:
            self.fitness = 1 / float(self.path_distance())
        return self.fitness
```

```python
# Creating a random path from a list of vertices
def create_path(vertices):
    path = random.sample(vertices, len(vertices))
    return path
```

```python
# Creating number of random paths equal to population_size
def initial_population_genetarion(population_size, vertices):
    population = []

    for i in range(0, population_size):
        population.append(create_path(vertices))
        
    return population
```

```python
# Sorting of population by fitness DESC
def rank_paths_by_fitness(population):
    fitness_results = {}
    
    for i in range(0, len(population)):
        fitness_results[i] = Fitness(population[i]).path_fitness()
        
    return sorted(fitness_results.items(), key=operator.itemgetter(1), reverse=True)
```

```python
# Sorting of population by distance ASC
def rank_paths_by_distance(population):
    fitness_results = {}
    
    for i in range(0, len(population)):
        fitness_results[i] = Fitness(population[i]).path_distance()
        
    return sorted(fitness_results.items(), key=operator.itemgetter(1), reverse=False)
```

```python
# Selection method via roulette wheel
def selection(population_ranked, elite_size):
    selection_results = []
    df = pd.DataFrame(np.array(population_ranked), columns=["Index", "Fitness"])
    df['cumulative_sum'] = df.Fitness.cumsum()
    df['cumulative_percent'] = 100 * df.cumulative_sum / df.Fitness.sum()

    for i in range(0, elite_size):
        selection_results.append(population_ranked[i][0])
        
    for i in range(0, len(population_ranked) - elite_size):
        pick = 100 * random.random()
        for i in range(0, len(population_ranked)):
            if pick <= df.iat[i, 3]:
                selection_results.append(population_ranked[i][0])
                break
                
    return selection_results
```

```python
# Creating a list of selected parents
def create_mating_pool(population, selection_results):
    mating_pool = []
    
    for i in range(0, len(selection_results)):
        index = selection_results[i]
        mating_pool.append(population[index])
        
    return mating_pool
```

```python
# Crossover of two parent objects via choosen method
def crossover(parent1, parent2):
    random_crossover = 0
    
    if crossover_type == "RANDOM":
        random_crossover = random.randint(1, 3)
    if crossover_type == "PMX" or random_crossover == 1:
        child = PMX_crossover(parent1, parent2)
    elif crossover_type == "OX" or random_crossover == 2:
        child = OX_crossover(parent1, parent2)
    elif crossover_type == "CX" or random_crossover == 3:
        child = CX_crossover(parent1, parent2)
        
    return child
```

```python
# 2-opt algorithm used to improve the score of selected path
def two_opt(path):
    best_path = path
    best_path_len = -1
    n = len(path)
    while path_distance_using_LUT(path) != best_path_len:
        best_path_len = path_distance_using_LUT(best_path)
        for i in range(1, n - 2):
            for j in range(i + 1, n + 1):
                if j - i == 1:
                    continue  
                new_path = path[:]
                new_path[i:j] = path[j - 1:i - 1:-1] 
                if path_distance_using_LUT(new_path) < path_distance_using_LUT(best_path):
                    best_path = new_path
        path = best_path
        
    return best_path
```

```python
# 2-opt algorithm used to improve the score of selected path (short version)
def two_opt_short(path):
    best_path = path
    best_path_len = -1
    n = len(path)
    best_path_len = path_distance_using_LUT(best_path)
    for i in range(1, n - 2):
        for j in range(i + 1, n + 1):
            if j - i == 1:
                continue  
            new_path = path[:]
            new_path[i:j] = path[j - 1:i - 1:-1] 
            if path_distance_using_LUT(new_path) < path_distance_using_LUT(best_path):
                best_path = new_path
        path = best_path
        
    return best_path
```

```python
# Usage of 2-opt algorithm on every member of population
def use_two_opt_on_population(population):
    new_population = []
    
    for i in range(len(population)):
        new_population.append(two_opt_short(population[i]))
        
    return new_population
```

```python
# Crossover PMX implementation
def PMX_crossover(parent1, parent2):
    p = [None] * len(parent1)
    sub1 = []
    sub2 = []

    # selecting segment to copy from parent
    cut_a = int(random.random() * len(parent1))
    cut_b = int(random.random() * len(parent1))
    start = min(cut_a, cut_b)
    end = max(cut_a, cut_b)

    # copying segment
    for i in range(start, end):
        p[i] = parent2[i]
        sub1.append(parent1[i])
        sub2.append(parent2[i])

    # copy vertices that are not in child object already
    for i in range(0, len(parent1)):
        if parent1[i] not in p and p[i] is None:
            p[i] = parent1[i]

    # copying rest of the vertices to their corresponding places
    for i in range(0, len(parent1)):
        if p[i] is None:
            s1 = sub1.copy()
            s2 = sub2.copy()
            if parent1[i] in s1:
                vertex = s2[s1.index(parent1[i])]
                index = s2.index(vertex)
            else:
                vertex = s1[s2.index(parent1[i])]
                index = s1.index(vertex)
            del s1[index]
            del s2[index]
            while vertex in p:
                if vertex in s1:
                    vertex = s2[s1.index(vertex)]
                    index = s2.index(vertex)
                else:
                    vertex = s1[s2.index(vertex)]
                    index = s1.index(vertex)
                del s1[index]
                del s2[index]
            p[i] = vertex

    return p
```

```python
# Crossover OX implementation
def OX_crossover(parent1, parent2):
    p1 = []

    # selecting segment to copy from parent
    cut_a = int(random.random() * len(parent1))
    cut_b = int(random.random() * len(parent1))
    start = min(cut_a, cut_b)
    end = max(cut_a, cut_b)

    # copying segment
    for i in range(start, end):
        p1.append(parent1[i])

    # selecting rest of the vertices in order
    p2 = [item for item in parent2 if item not in p1]

    # combining copied segment and rest of vertices
    child = p1 + p2
    
    return child
```

```python
# Crossover CX (cycle crossover) implementation
def CX_crossover(parent1, parent2):
    p = [None] * len(parent1)
    
    # selecting first vertex
    current_index = 0
    p[0] = parent1[0]

    # finding a cycle in parents starting from first vertex
    while parent2[current_index] not in p:
        p[parent1.index(parent2[current_index])] = parent2[current_index]
        current_index = parent1.index(parent2[current_index])

    # copying rest of the vertices to their corresponding places
    for i in range(0, len(parent1)):
        if p[i] is None:
            p[i] = parent2[i]

    return p
```

```python
# Performing crossover on all but top elite_size objects in population
def breed_population(mating_pool, elite_size):
    children = []
    length = len(mating_pool) - elite_size
    pool = random.sample(mating_pool, len(mating_pool))

    for i in range(0, elite_size):
        children.append(mating_pool[i])

    for i in range(0, length):
        chance = random.uniform(0, 1)
        if chance < crossover_rate:
            child = crossover(pool[i], pool[len(mating_pool) - i - 1])
        else:
            child = pool[i]
        children.append(child)
        
    return children
```

```python
# Mutaton of single object (individual) by swapping vertices
def mutate(individual):
    for swapped in range(len(individual)):
        if random.random() < mutation_rate:
            swap_with = int(random.random() * len(individual))

            vertex1 = individual[swapped]
            vertex2 = individual[swap_with]

            individual[swapped] = vertex2
            individual[swap_with] = vertex1
            
    return individual
```

```python
# Mutating all members of population 
def mutate_population(population):
    mutated_population = []

    for i in range(0, len(population)):
        mutated = mutate(population[i])
        mutated_population.append(mutated)
        
    return mutated_population
```

```python
# Creating next generation by ranking individuals in population, selecting and crossing them, mutating and optimizing them
def next_generation(current_generation, elite_size):
    ranked_population = rank_paths_by_fitness(current_generation)
    selected = selection(ranked_population, elite_size)
    mating_pool = create_mating_pool(current_generation, selected)
    children = breed_population(mating_pool, elite_size)
    next_gen = mutate_population(children)
    
    if two_opt_enabled:
        next_gen = use_two_opt_on_population(next_gen)
        
    return next_gen
```

```python
# Main function responsible for generic algorithm (with plot generation and additional info)
def genetic_algorithm_with_plot(pop, pop_size, elite_size, generations, max_x, max_y):
    # creating initial population
    population = initial_population_genetarion(pop_size, pop)
    
    # printing shortest distance in initial population
    if show_full_info:
        print("Shortest distance in initial population: " + str(1 / rank_paths_by_fitness(population)[0][1]))
        print("Longest distance in initial population: " + str(1 / rank_paths_by_fitness(population)[len(population) - 1][1]))
    
    # calculating fitness for initial population
    fitness_progress = []
    fitness_progress.append(1 / rank_paths_by_fitness(population)[0][1])

    # going through each generation
    for i in range(0, generations):
        population = next_generation(population, elite_size)
        fitness_progress.append(1 / rank_paths_by_fitness(population)[0][1])
        print("Generation " + str(i) + " in progress...") 

    if show_full_info:
        ranked_paths = rank_paths_by_distance(population)
        best_path_index = ranked_paths[0][0]
        best_path_distance = ranked_paths[0][1]
        worst_path_distance = ranked_paths[len(population) - 1][1]

        plot_data = []
        for j in range(0, len(population[best_path_index])):
            plot_data.append((population[best_path_index][j].getX(), population[best_path_index][j].getY()))

        plot_data.append((population[best_path_index][0].getX(), population[best_path_index][0].getY()))
        
        print('Shortest distance in final population: ' + str(best_path_distance))
        print('Longest distance in final population: ' + str(worst_path_distance))
        
        for fit in fitness_progress:
            print(fit)

        # plot visualising shortest path found by algorithm
        plt.plot([x[0] for x in plot_data],
                 [x[1] for x in plot_data],
                 color='black', lw=1)
        plt.ylabel('y')
        plt.xlabel('x')
        plt.xlim(0, max_x * 1.1)
        plt.ylim(0, max_y * 1.1)
        plt.title('Shortest path in ' + str(generations) + ' generations')
        plt.show()

        # plot showing how shortest path distance changed over generations
        plt.plot(fitness_progress)
        plt.ylabel('Distance')
        plt.xlabel('Generation')
        plt.title('Shortest distance over generations')
        plt.show()
    else:
        print("Shortest distance in final population: " + str(1 / rank_paths_by_fitness(population)[0][1]))
```

```python
# Function responsible for generic algorithm (returning only final result)
def genetic_algorithm_silent(pop, pop_size, elite_size, generations, fitness_progress):
    # creating initial population
    population = initial_population_genetarion(pop_size, pop)

    # going through each generation
    for i in range(0, generations):
        population = next_generation(population, elite_size)
        fitness_progress[i] += (1 / rank_paths_by_fitness(population)[0][1])
    
    return 1 / rank_paths_by_fitness(population)[0][1]
```

```python
# Distance between two vertices
def distance(vertex1, vertex2):
    x_distance = abs(vertex1.x - vertex2.x)
    y_distance = abs(vertex1.y - vertex2.y)
    distance = np.sqrt((x_distance ** 2) + (y_distance ** 2))
    
    return distance
```

```python
# Calculating path distance using LUT to improve time complexity 
def path_distance_using_LUT(path):
    dist = 0
    
    for i in range(0, len(path) - 1):
        dist += distanceLUT[path[i], path[i+1]]
        
    return dist
```

```python
# Generating LUT table for distances between vertices
def create_distance_LUT(vertex_list):
    distanceLUT = {}
    
    for i in range(0, len(vertex_list)):
        for j in range(0, len(vertex_list)):
            if i != j:
                distanceLUT[vertex_list[i], vertex_list[j]] = distance(vertex_list[i], vertex_list[j])
            else:
                distanceLUT[vertex_list[i], vertex_list[j]] = None
                
    return distanceLUT
```

```python
# Creating list of vertices where path will be looked for
def get_list_of_vertices():
    vertex_list = []
    
    if read_from_file:
        infile = open(file_path, 'r')
        listOfLines = infile.readlines()
        max_x = 0
        max_y = 0
        ignore_lines = 0
        if file_path[-4:] == ".tsp":
            ignore_lines = 6
            listOfLines = listOfLines[:-1]
        # iterate over the lines in file
        for line in listOfLines:
            if ignore_lines > 0:
                ignore_lines = ignore_lines - 1
            else:
                vertex_list.append(Vertex(x=float(line.split()[1]), y=float(line.split()[2])))
                if float(line.split()[1]) > max_x:
                    max_x = float(line.split()[1])
                if float(line.split()[2]) > max_y:
                    max_y = float(line.split()[2])
        infile.close()
    else:
        max_x = max_x_to_generate
        max_y = max_y_to_generate
        # generate new vetrices
        for i in range(0, vertices_number):
            vertex_list.append(Vertex(x=int(random.random() * max_x), y=int(random.random() * max_y)))
    
    return vertex_list, max_x, max_y
```

```python
# show plot and score for initial population only
def initial_population_statistics(vertex_list):
    initial_path_distance = path_distance_using_LUT(vertex_list)
    print('Path distance: ' + str(initial_path_distance))
    
    plot_data = []
    for i in range(0, len(vertex_list)):
        plot_data.append((vertex_list[i].getX(), vertex_list[i].getY()))
    plot_data.append((vertex_list[0].getX(), vertex_list[0].getY()))

    # plot visualising shortest path found by algorithm
    plt.plot([x[0] for x in plot_data],
             [x[1] for x in plot_data],
             color='black', lw=1)
    plt.ylabel('y')
    plt.xlabel('x')
    plt.xlim(0, max_x * 1.1)
    plt.ylim(0, max_y * 1.1)
    plt.title('Initial path')
    plt.show()
    
    new_vertex_list = two_opt(vertex_list)
    
    opt_path_distance = path_distance_using_LUT(new_vertex_list)
    print('Path distance after 2-opt: ' + str(opt_path_distance))
    
    plot_data = []
    for i in range(0, len(new_vertex_list)):
        plot_data.append((new_vertex_list[i].getX(), new_vertex_list[i].getY()))
    plot_data.append((new_vertex_list[0].getX(), new_vertex_list[0].getY()))

    # plot visualising shortest path found by algorithm
    plt.plot([x[0] for x in plot_data],
             [x[1] for x in plot_data],
             color='black', lw=1)
    plt.ylabel('y')
    plt.xlabel('x')
    plt.xlim(0, max_x * 1.1)
    plt.ylim(0, max_y * 1.1)
    plt.title('Initial path after 2-opt')
    plt.show()
    
    print('Path was shorter by : ' + str(100 - 100 * opt_path_distance/initial_path_distance) + '%')
```

# MAIN

```python
# getting list of vertices
random.seed(generator_seed)
vertex_list, max_x, max_y = get_list_of_vertices()

# creating lookup table for vetrex distances
distanceLUT = create_distance_LUT(vertex_list)

random.seed(algorithm_seed)
if test_initial_population_only:
    initial_population_statistics(vertex_list)
else:
    # initializing genetic algorithm
    genetic_algorithm_with_plot(
        pop=vertex_list, 
        pop_size=population_size, 
        elite_size=elite_size, 
        generations=generations, 
        max_x=max_x, 
        max_y=max_y)
```

```python
if do_mean_check and not test_initial_population_only:
    results = []
    sum = 0.0
    for i in range(0, get_mean_from):
        print('Iteration '+ str(i) + ' in progress...')
        
        fitness_progress = [0] * generations

        current = genetic_algorithm_silent(
            pop=vertex_list, 
            pop_size=population_size, 
            elite_size=elite_size, 
            generations=generations, 
            fitness_progress=fitness_progress)
        results.append(current)
        sum += current
        
        if show_full_info:
            for fit in fitness_progress:
                print(fit)
    
    print("")
    print('Results:')
    print(results)
    print("")
    print('Mean from ' + str(get_mean_from) + ' is '+ str(sum/get_mean_from))
```
